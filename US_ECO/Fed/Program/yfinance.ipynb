```python
import yfinance as yf

def get_google_stock_data():
    """
    Retrieves Google (GOOG) stock price and fundamental data using yfinance.

    Returns:
        A dictionary containing the stock price and fundamental data, 
        or None if an error occurs.
    """

    try:
        # Create a Ticker object for Google (GOOG)
        google = yf.Ticker("GOOG")  # You can use "GOOGL" for Google's Class A shares

        # Get the current stock price
        current_price = google.fast_info.last_price

        # Get fundamental data
        fundamental_data = google.info  # This provides a lot of data!

        # Extract some key fundamental data (customize as needed)
        key_fundamentals = {
            "market_cap": fundamental_data.get("marketCap", None),  # Use .get() for safe access
            "price_to_earnings_ratio": fundamental_data.get("trailingPE", None),
            "revenue": fundamental_data.get("revenue", None),
            "debt_to_equity_ratio": fundamental_data.get("debtToEquity", None),
            "dividend_yield": fundamental_data.get("dividendYield", None),
            "current_price": current_price
        }


        return key_fundamentals

    except Exception as e:
        print(f"An error occurred: {e}")
        return None



if __name__ == "__main__":
    data = get_google_stock_data()

    if data:
        print("Google (GOOG) Stock Data:")
        for key, value in data.items():
            print(f"{key}: {value}")
    else:
        print("Failed to retrieve stock data.")
```
