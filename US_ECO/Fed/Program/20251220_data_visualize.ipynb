```python

import pandas as pd
import matplotlib.pyplot as plt

def plot_data_from_csv(csv_file, x_column, y_column, 
                       title="Data Plot", xlabel="X-axis", 
                       ylabel="Y-axis", plot_type='line'):
    """
    Plots data from a CSV file using Matplotlib.

    Args:
        csv_file (str): Path to the CSV file.
        x_column (str): Name of the column to use for the x-axis.
        y_column (str): Name of the column to use for the y-axis.
        title (str): Title of the plot (default: "Data Plot").
        xlabel (str): Label for the x-axis (default: "X-axis").
        ylabel (str): Label for the y-axis (default: "Y-axis").
        plot_type (str): Type of plot to create ('line', 'scatter', 'bar'). Default 'line'.
    """

    try:
        # Read the CSV file into a Pandas DataFrame
        df = pd.read_csv(csv_file)

        # Check if the specified columns exist
        if x_column not in df.columns or y_column not in df.columns:
            raise ValueError("One or both specified columns not found in CSV file.")
        
        # Create the plot based on the specified plot type
        plt.figure(figsize=(10, 6))  # Adjust figure size for better readability

        if plot_type == 'line':
            plt.plot(df[x_column], df[y_column], marker='o')  # Add markers for better visibility
        elif plot_type == 'scatter':
            plt.scatter(df[x_column], df[y_column])
        elif plot_type == 'bar':
            plt.bar(df[x_column], df[y_column]) #requires x_column to contain categorical data 
        else:
            raise ValueError("Invalid plot_type. Choose 'line', 'scatter', or 'bar'.")

        # Add labels and title
        plt.title(title)
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)

        # Rotate x-axis labels for better readability if needed (for categorical data)
        plt.xticks(rotation=45, ha='right')  # Adjust rotation for readability
        plt.tight_layout() #Adjust plot to fit labels
        # Display the plot
        plt.show()

    except FileNotFoundError:
        print(f"Error: CSV file not found at {csv_file}")
    except ValueError as e:
        print(f"Error: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")



if __name__ == '__main__':
    # Example Usage:

    # 1. Create a sample CSV file (for demonstration)
    import csv
    data = [
        ['Year', 'Sales'],
        [2018, 100],
        [2019, 120],
        [2020, 150],
        [2021, 130],
        [2022, 160]
    ]
    with open('sales_data.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(data)



    # 2. Call the function to plot the data
    plot_data_from_csv(
        csv_file='sales_data.csv',
        x_column='Year',
        y_column='Sales',
        title='Sales Over Time',
        xlabel='Year',
        ylabel='Sales (in Units)',
        plot_type='line'  # Can be 'line', 'scatter', or 'bar'
    )

    # Example using scatter plot:
    # plot_data_from_csv(
    #     csv_file='sales_data.csv',
    #     x_column='Year',
    #     y_column='Sales',
    #     title='Sales Over Time (Scatter)',
    #     xlabel='Year',
    #     ylabel='Sales (in Units)',
    #     plot_type='scatter'
    # )

    # Example using bar plot:
    # plot_data_from_csv(
    #     csv_file='sales_data.csv',
    #     x_column='Year',
    #     y_column='Sales',
    #     title='Sales Over Time (Bar)',
    #     xlabel='Year',
    #     ylabel='Sales (in Units)',
    #     plot_type='bar'
    # )
```
