```python
from __future__ import annotations # this is to add the annotation of the code

import datetime
import csv
import sys
import time

import yfinance as yf
from datetime import datetime, timezone
from typing import Dict, Optional


NICKNAMES: Dict[str,str] = { # this is the annotation NICKNAME: Dict[str, str]
    "google": "GOOGL",
    "alphabet": "GOOGL",
    "apple": "AAPL",
    "intel": "INTC",
    "arm": "ARM",
    "qualcomm": "QCOM",
    "microsoft": "MSFT",
    "amd": "AMD",
    "nvidia": "NVDA",
    "bmo500": "ZSP",
    "qqq": "QQQ",
    "sqqq": "SQQQ",
    "spy500": "SPY",
    "gold": "AAAU",
    "amazon": "AMZN",
}


def resolve_ticker(user_input: str) -> str:
    key = user_input.strp().lower()

    return NICKNAMES.get(key, user_input.strip().upper())

def utc_now_iso() -> str:
    return datetime.now(timezone.utc).isoformat()


def fetch_quote(name:str) -> Optional[dict]:
    '''
    fetch the stock or the list data
    Optional indicates that the function may return none if nothing is fetched
    '''

    t = yf.ticker(name)
    
    # Try to use fast_info (fastest, minimal overhead)

    last = bid = ask = prev_close = volume = None
    currency = exchange = short_name = market_state = None 

    # the info contains more information
    try:
        fi = t.info
        # Some fields may be missing per ticker
        last = fi.get("last_price", None)
        bid = fi.get("bid", None)
        ask = fi.get("ask", None)
        bid_size = fi.get("bidSize", None)
        ask_size = fi.get("askSize", None)
        
        current_price = fi.get("currentPrice", None)
        target_high_price = fi.get("targetHighPrice", None)
        target_low_price = fi.get("targetLowPrice", None)
        target_mean_price = fi.get("targetMeanPrice", None)
        target_median_price = fi.get("targetMedianPrice", None)
        recommendation_key = fi.get("recommendationKey", None)
        number_of_analyst_opinions = fi.get("numberOfAnalystOpinions", None)
        
        last_dividend_value = fi.get("lastDividendValue", None)
        last_dividend_date = fi.get("lastDividendDate", None)

        total_cash = fi.get("totalCash", None)
        total_cash_per_share = fi.get("totalCashPerShare", None)
        ebitda = fi.get("ebitda", None)
        total_debt = fi.get("totalDebt", None)
        quick_ratio = fi.get("quickRatio", None)
        current_ratio = fi.get("currentRatio", None)
        total_revenue = fi.get("totalRevenue", None)
        debt_to_equity = fi.get("debtToEquity", None)
        revenue_per_share = fi.get("revenuePerShare", None)
        return_on_assets = fi.get("returnOnAssets", None)
        return_on_equity = fi.get("returnOnEquity", None)
        gross_profits = fi.get("grossProfits", None)
        free_cashflow = fi.get("freeCashflow", None)
        operating_cashflow = fi.get("operatingCashflow", None)
        earnings_growth = fi.get("earningsGrowth", None)
        revenue_growth = fi.get("revenueGrowth", None)
        gross_margins = fi.get("grossMargins", None)
        ebitda_margins = fi.get("ebitdaMargins", None)
        operating_margins = fi.get("operatingMargins", None)

        regular_market_change = fi.get("regularMarketChange", None)
        regular_market_change_percent = fi.get("regularMarketChangePercent", None)
        regular_market_price = fi.get("regularMarketPrice", None)

        post_market_change_percent = fi.get("postMarketChangePercent", None)
        post_market_price = fi.get("postMarketPrice", None)
        post_market_change = fi.get("postMarketChange", None)
        regular_market_change = fi.get("regularMarketChange", None)

        average_daily_volume_3_month = fi.get("averageDailyVolume3Month", None)

        open_price = fi.get("open", None)
        day_low = fi.get("dayLow", None)
        day_high = fi.get("dayHigh", None)
        regular_market_previous_close = fi.get("regularMarketPreviousClose", None)
        regular_market_open = fi.get("regularMarketOpen", None)
        regular_market_day_low = fi.get("regularMarketDayLow", None)
        regular_market_day_high = fi.get("regularMarketDayHigh", None)
        dividend_rate = fi.get("dividendRate", None)
        dividend_yield = fi.get("dividendYield", None)

        trail_PE = fi.get("trailingPE", None)
        forward_PE = fi.get("forwardPE", None)

        trail_Eps = fi.get("trailingEps", None) or fi.get("epsTrailing", None)
        forward_Eps = fi.get("forwardEps", None) or fi.get("epsForward", None)

        
        prev_close = fi.get("previous_close", None)
        volume = fi.get("last_volume", None) or fi.get("volume", None)
        reg_market_volume = fi.get("regularMarketVolume", None)
        avg_volume = fi.get("averageVolume", None)
        ave_10d_volume = fi.get("averageVolume10days", None)

        currency = fi.get("currency", None)
        exchange = fi.get("exchange", None)
        
        profit_margin = fi.get("profitMargins", None)
        float_shares = fi.get("floatShares", None)
        company_value = fi.get("enterpriseValue", None)

        float_share = fi.get("floatShares", None)
        outstanding_share = fi.get("sharesOutstanding", None)
        share_short = fi.get("sharesShort", None)

        share_percent_shares_out = fi.get("sharesPercentSharesOut", None)

        insider_held_percent = fi.get("heldPercentInsiders", None)
        institution_held_percent = fi.get("heldPercentInstitutions", None)

        short_ratio = fi.get("shortRatio", None)
        short_percent_of_float = fi.get("shortPercentOfFloat", None)
        implied_shares_outstanding = fi.get("impliedSharesOutstanding", None)

        book_value = fi.get("bookValue", None)
        price_2_book = fi.get("priceToBook", None)






        
        
    

    except Exception:
        pass
    

    try:
        info = getattr(t, "info", {}) or {}

    except Exception:
        pass
```
